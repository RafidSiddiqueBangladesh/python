Dataset Description
1.Context or Background

      This dataset contains medical records from patients, aiming to predict the presence of       heart disease based on several health-related features.
      The objective is to classify whether a patient has heart disease (1) or not (0) based        on their medical information.
2.Number of Rows and Columns

     Rows: 303 (patient records)
     Columns: 14 (features including the target variable)
3.Explanation of Key Columns

      age: Age of the patient (in years)
      sex: Gender (1 = Male, 0 = Female)
      cp: Chest pain type (categorical: 0–3)
      trestbps: Resting blood pressure (in mm Hg)
      chol: Serum cholesterol level (in mg/dL)
      fbs: Fasting blood sugar (> 120 mg/dL, 1 = True, 0 = False)
      restecg: Resting electrocardiographic results (values 0, 1, 2)
      thalach: Maximum heart rate achieved
      exang: Exercise-induced angina (1 = Yes, 0 = No)
      oldpeak: ST depression induced by exercise relative to rest
      slope: Slope of the peak exercise ST segment
      ca: Number of major vessels (0–3) colored by fluoroscopy
      thal: Thalassemia (3 = Normal, 6 = Fixed defect, 7 = Reversible defect)
      target: Presence of heart disease (1 = Yes, 0 = No)

```python
2️⃣ Distribution and Relationship Plots (40 Marks)
Create the following visualizations using Seaborn:
2.1Histogram/Distplot: Display the distribution of a numeric variable. (10 Marks)
2.2Pairplot: Show pairwise relationships between numerical variables. (10 Marks)
2.3Heatmap: Visualize correlations between numerical features. (10 Marks)
2.4Jointplot: Analyze the relationship between two variables, including marginal distributions. (10 Marks)
```

```python
df = pd.read_csv("/kaggle/input/heart-disease-cleveland/Heart_disease_cleveland_new.csv")

# Display first 5 rows
df.head()
```

```python
#2.1 Histogram / Distplot (10 Marks)



plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color="skyblue")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Patients")
plt.show()
```

```python
#2.2 Pairplot (10 Marks)
sns.pairplot(df[['age', 'chol', 'trestbps', 'thalach', 'target']], hue='target', diag_kind='kde')
plt.show()
```

```python
#3.3 Heatmap (10 Marks)



plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#3.4 Jointplot (10 Marks
sns.jointplot(x='age', y='chol', data=df, kind='reg', color="purple")
plt.show()
```

3️⃣ Categorical Data Visualization (20 Marks)
Use Seaborn to analyze and visualize categorical data:
  3.1 Bar Plot: Compare categories using aggregation. (10 Marks)
  3.2 Box Plot/Violin Plot: Visualize the distribution of a numeric variable across categories. (10 Marks)


```python
#3.1 Barplot


plt.figure(figsize=(8, 5))
sns.barplot(x='target', y='chol', data=df, estimator=lambda x: sum(x)/len(x), palette="pastel")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Average Cholesterol")
plt.title("Average Cholesterol by Heart Disease Status")
plt.show()
```

```python
#3.2 Box plot


plt.figure(figsize=(8, 5))
sns.boxplot(x='cp', y='age', data=df, palette="coolwarm")
plt.xlabel("Chest Pain Type")
plt.ylabel("Age")
plt.title("Age Distribution by Chest Pain Type")
plt.show()
```
