1. Grouping and Aggregation (20 Marks)
You are provided with the following dataset about employees and their sales performance:
“”””import pandas as pd

data = {
    'Employee': ['Alice', 'Bob', 'Alice', 'Bob', 'Charlie', 'Alice', 'Charlie', 'Bob'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'North', 'North'],
    'Sales': [200, 150, 300, 400, 100, 250, 300, 350],
    'Year': [2023, 2023, 2024, 2024, 2023, 2024, 2024, 2023]
}

df = pd.DataFrame(data) “”””

Perform the following tasks:
1.Group the data by Employee and calculate the total sales for each employee. (5 Marks)
2.Group the data by Region and find the average sales per region. (5 Marks)
3.Group the data by both Employee and Year to calculate the total sales for each combination. (10 Marks)


```python
import pandas as pd

data = {
    'Employee': ['Alice', 'Bob', 'Alice', 'Bob', 'Charlie', 'Alice', 'Charlie', 'Bob'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'North', 'North'],
    'Sales': [200, 150, 300, 400, 100, 250, 300, 350],
    'Year': [2023, 2023, 2024, 2024, 2023, 2024, 2024, 2023]
}

df = pd.DataFrame(data)
#1. Group by Employee and Calculate Total Sales (5 Marks)
employee_sales = df.groupby('Employee')['Sales'].sum()
print("Total Sales by Employee:\n", employee_sales)


#2. Group by Region and Find Average Sales per Region (5 Marks)
region_avg_sales = df.groupby('Region')['Sales'].mean()
print("Average Sales by Region:\n", region_avg_sales)

#3. Group by Both Employee and Year to Calculate Total Sales for Each Combination (10 Marks)
employee_year_sales = df.groupby(['Employee', 'Year'])['Sales'].sum()
print("Total Sales by Employee and Year:\n", employee_year_sales)


```

2. Merging and Concatenation (20 Marks)
You are provided with two datasets containing employee details and sales targets:
employee_details = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Department': ['Sales', 'Sales', 'Marketing', 'HR']
})

sales_targets = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie', 'David'],
    'Sales_Target': [1000, 1200, 800, 500]
})

Perform the following tasks:
1.Merge the two datasets on the Employee column to combine all details. (10 Marks)
2.Create a new column showing the difference between Sales_Target and total sales (calculated from Task 1). Hint: Use the merged dataset for calculations. (10 Marks)


```python
import pandas as pd


employee_details = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Department': ['Sales', 'Sales', 'Marketing', 'HR']
})

sales_targets = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie', 'David'],
    'Sales_Target': [1000, 1200, 800, 500]
})

# 1Merging 10marks
merged_data = pd.merge(employee_details, sales_targets, on='Employee')
print("Merged Dataset:\n", merged_data)



# 2 Create a New Column Showing the Difference Between Sales_Target and Total Sales (10 Marks)
total_sales = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie'],
    'Total_Sales': [750, 900, 400]
})


final_data = pd.merge(merged_data, total_sales, on='Employee', how='left')
final_data['Total_Sales'] = final_data['Total_Sales'].fillna(0)
final_data['Sales_Difference'] = final_data['Sales_Target'] - final_data['Total_Sales']

print("Final Dataset with Sales Difference:\n", final_data)

```

The following dataset contains daily stock prices:
stock_prices = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Price': [100, 102, 105, 107, 110]
})

Perform the following tasks:
1 Convert the Date column to a datetime object. (5 Marks)
2 Resample the data to calculate the weekly average price. (10 Marks)
3 Find the day with the highest stock price. (5 Marks)


```python
import pandas as pd

# Provided dataset
stock_prices = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Price': [100, 102, 105, 107, 110]
})
#1. Convert the Date Column to a Datetime Object (5 Marks)
stock_prices['Date'] = pd.to_datetime(stock_prices['Date'])
print("Dataset with Datetime Column:\n", stock_prices)
#2. Resample the Data to Calculate the Weekly Average Price (10 Marks)

stock_prices.set_index('Date', inplace=True)


weekly_avg_price = stock_prices.resample('W').mean()
print("Weekly Average Price:\n", weekly_avg_price)
#3. Find the Day with the Highest Stock Price (5 Marks)

stock_prices.reset_index(inplace=True)


highest_price_day = stock_prices.loc[stock_prices['Price'].idxmax()]
print("Day with the Highest Stock Price:\n", highest_price_day)

```

4. Pivot Tables and Cross-Tabulations (20 Marks)
Using the original df dataset from Task 1:
1 Create a pivot table to find the total sales by Region and Year. (10 Marks)
2 Create a cross-tabulation showing the count of employees working in each Region for each Year. (10 Marks)


```python

import pandas as pd

data = {
    'Employee': ['Alice', 'Bob', 'Alice', 'Bob', 'Charlie', 'Alice', 'Charlie', 'Bob'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'North', 'North'],
    'Sales': [200, 150, 300, 400, 100, 250, 300, 350],
    'Year': [2023, 2023, 2024, 2024, 2023, 2024, 2024, 2023]
}

df = pd.DataFrame(data)
#1. Create a Pivot Table to Find Total Sales by Region and Year (10 Marks)

pivot_table_sales = df.pivot_table(values='Sales', index='Region', columns='Year', aggfunc='sum', fill_value=0)

print("Pivot Table: Total Sales by Region and Year\n", pivot_table_sales)
#2. Create a Cross-Tabulation for the Count of Employees in Each Region for Each Year (10 Marks)

cross_tab_employees = pd.crosstab(index=df['Region'], columns=df['Year'], values=df['Employee'], aggfunc='count', margins=False)

print("Cross-Tabulation: Count of Employees by Region and Year\n", cross_tab_employees)


```
