2. Basic Plots (40 Marks)
Use Matplotlib to create the following visualizations from your chosen dataset:
1 Line Plot: Show a trend or progression of a variable over time or another index. (10 Marks)
2 Bar Chart: Compare categories or groups. (10 Marks)
3 Scatter Plot: Illustrate relationships between two numeric variables. (10 Marks)
4 Box Plot: Summarize the distribution and identify outliers of a numeric variable. (10 Marks)


```python
2.1#line plot 10 marks
import matplotlib.pyplot as plt

# Group by Year and sum the consumption for a specific meat type  Poultry
yearly_poultry_consumption = df.groupby("Year")["Poultry"].sum()

# Line plot for Poultry consumption over time
plt.figure(figsize=(10, 5))
plt.plot(yearly_poultry_consumption.index, yearly_poultry_consumption.values, marker="o", linestyle="-", color="blue")
plt.xlabel("Year")
plt.ylabel("Poultry Consumption (kg per capita)")
plt.title("Poultry Consumption Over Time")
plt.grid()
plt.show()

```

```python
#2.2 Bar chart 10marks
# Group by Year and sum the consumption for different meats
yearly_meat_consumption = df.groupby("Year")[["Poultry", "Beef", "Fish and seafood"]].sum()

# Create the bar chart
yearly_meat_consumption.plot(kind="bar", figsize=(16, 8), edgecolor="black", color=["skyblue", "lightgreen", "lightcoral"])

# Labels and title
plt.xlabel("Year")
plt.ylabel("Total Consumption (kg per capita)")
plt.title("Meat Consumption by Type Over Time")

plt.legend(title="Meat Type")
plt.show()
```

```python
#2.3 scatter 10marks
# Scatter plot to show relationship between Beef and Poultry consumption
plt.figure(figsize=(8, 5))
plt.scatter(df["Poultry"], df["Beef"], alpha=0.5, color="purple")

# Labels and title
plt.xlabel("Poultry Consumption (kg per capita)")
plt.ylabel("Beef Consumption (kg per capita)")
plt.title("Relationship Between Poultry and Beef Consumption")
plt.grid(True)
plt.show()
```

```python
#2.4 boxplot 10marks
# Box plot for Poultry consumption
plt.figure(figsize=(8, 5))
df.boxplot(column="Poultry")

# Title
plt.title("Box Plot of Poultry Consumption")
plt.ylabel("Poultry Consumption (kg per capita)")
plt.show()
```

3. Customizing Plots (20 Marks)
 1 Add meaningful titles, labels, and legends to all the plots. (10 Marks)
 2  Experiment with colors and markers to make your plots more readable and visually appealing. (10 Marks)


```python
import matplotlib.pyplot as plt

#3.1 Group by Year and sum poultry consumption
yearly_poultry_consumption = df.groupby("Year")["Poultry"].sum()

# Create a customized line plot
plt.figure(figsize=(10, 5))
plt.plot(yearly_poultry_consumption.index, yearly_poultry_consumption.values, 
         marker="o", linestyle="-", color="darkorange", linewidth=2, markersize=6)

# Customization
plt.xlabel("Year", fontsize=12, fontweight="bold")
plt.ylabel("Poultry Consumption (kg per capita)", fontsize=12, fontweight="bold")
plt.title(" Poultry Consumption Over Time", fontsize=14, fontweight="bold", color="darkred")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
3.2# Group by Year and sum the consumption for different meat types
yearly_meat_consumption = df.groupby("Year")[["Poultry", "Beef", "Fish and seafood"]].sum()

# Create a bar chart with improved styling
plt.figure(figsize=(12, 6))
yearly_meat_consumption.plot(kind="bar", ax=plt.gca(), edgecolor="black", 
                             color=["#1f77b4", "#ff7f0e", "#2ca02c"])

# Customization
plt.xlabel("Year", fontsize=12, fontweight="bold")
plt.ylabel("Total Consumption (kg per capita)", fontsize=12, fontweight="bold")
plt.title("Meat Consumption by Type Over Time", fontsize=14, fontweight="bold", color="darkblue")
plt.xticks(rotation=45)
plt.legend(title="Meat Type", fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```

```python
#3.3 Create a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df["Poultry"], df["Beef"], alpha=0.5, color="purple", edgecolors="black")

# Customization
plt.xlabel("Poultry Consumption (kg per capita)", fontsize=12, fontweight="bold")
plt.ylabel("Beef Consumption (kg per capita)", fontsize=12, fontweight="bold")
plt.title("Relationship Between Poultry and Beef Consumption", fontsize=14, fontweight="bold", color="darkgreen")
plt.grid(True, linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#3.4 Create a box plot
plt.figure(figsize=(8, 5))
df.boxplot(column="Poultry", patch_artist=True, boxprops=dict(facecolor="lightblue"))

# Customization
plt.title("📊 Distribution of Poultry Consumption", fontsize=14, fontweight="bold", color="darkred")
plt.ylabel("Poultry Consumption (kg per capita)", fontsize=12, fontweight="bold")

# Show the plot
plt.show()
```

Insights and Observations from Visualizations (15 Marks)
After creating the plots, we need to interpret the findings and extract meaningful insights from each visualization.



1. Line Plot - Poultry Consumption Over Time
 Observation:

The poultry consumption trend shows a steady increase over time, indicating that poultry has become a more preferred source of meat.
There may be a rise in demand due to factors such as affordability, health benefits, or dietary preferences shifting away from red meat.
 Insight:

Poultry is the fastest-growing meat type, possibly due to economic and health-related factors.
Further investigation can determine if cultural preferences or government policies played a role in this increase.




2. Bar Chart - Meat Consumption Comparison by Year


 Observation:

Poultry and Fish & Seafood consumption have higher values compared to Beef.
Beef consumption appears lower and stable, whereas Poultry and Fish have grown over time.
Insight:

The shift toward poultry and fish could indicate that people prefer leaner and healthier meat options.
Beef consumption might be affected by cost, environmental concerns, or health awareness.





3. Scatter Plot - Relationship Between Poultry and Beef Consumption
Observation:

There is no strong correlation between Poultry and Beef consumption—higher poultry consumption doesn’t necessarily mean lower beef consumption.
Some regions may consume both, while others may favor one over the other.
Insight:

Poultry and beef consumption are likely influenced by economic factors, cultural preferences, or dietary habits.
Further analysis could check whether income levels or government regulations affect meat preferences.




 4. Box Plot - Distribution of Poultry Consumption
 Observation:

The median consumption is relatively high, meaning most regions consume a significant amount of poultry.
Some outliers indicate that certain regions have extremely high or low poultry consumption, which could be due to economic, religious, or regional dietary preferences.

Insight:

Some countries consume significantly more poultry than others, possibly due to availability, price, or food culture.
The outliers suggest regions where poultry consumption is extremely high or low, which may require further study to understand dietary trends and food policies.


Summary of Insights
1.Poultry consumption is rising, likely due to affordability and health preferences.
2.Beef consumption is relatively stable, possibly due to price or dietary concerns.
3.Poultry and beef consumption do not strongly correlate, showing that different factors influence their demand.
4.Outliers in poultry consumption suggest regional differences, requiring further investigation.
